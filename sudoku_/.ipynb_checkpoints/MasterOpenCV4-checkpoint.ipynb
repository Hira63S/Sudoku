## How to deal with image formatting in Matplotlib vs. OpenCV

```python
# Matplotlib uses RGB image structure whereas OpenCV uses GBR.

# load image using cv2.imread:
tesla_img = cv2.imread('tesla.png') # reads in and prints out an array.
print(tesla_img)
# we save it in tesla_img variable because cv2.imread loads it in bgr order.
```

```python
# since cv2.imread loads image in the BGR order, we need to split it into the three channels (b, g, r)
b, g, r = cv2.split(tesla_img)
```

```python
# now, we merge the channel again but in a different order to be used by matplotlib

tesla_plt = cv2.merge([r, g, b])
```

Now, we have two variables, tesla_img from cv2 and tesla_plt from matplot lib.
To plot the images with matplotlib in the same window, we use subplot:
You have three parameters to use within subplot, for example subplot(m,n,p). In this case, subplot handles plots in a m x n grid, where m establishes the number of rows, n establishes the number of columns, and p establishes where you want to place your plot in the grid. To show the images with Matplotlib, we will use imshow. 

```python
# we will show two image horizontally, m=1, n=2, and p=1
# shows image in the wrong color:
plt.subplot(121)
plt.imshow(tesla_img)

# Show image in the true color:
plt.subplot(122)
plt.imshow(tesla_plt)
plt.show()
```

```python
# now let's try printing images in the following way:

# this will show the image in true color:
cv2.imshow('bgr image', tesla_img)

# This will show it in wrong color:

cv2.imshow('rgb image', tesla_plt)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# If we want to show the images in the same window, we can use numpy's concatenate.
# to stack them horizontally, we will set axis=1
import numpy as np

img_concats = np.concatenate((tesla_img, tesla_plt), axis=1)
# the above function concatenates the image
# now we show the concatenated iamge:
cv2.imshow('bgr image and rgb image', img_concats)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# cv2.split() is a time-consuming operation.
# Instead, you can use numpy's indexing

B = tesla_img[:, :, 0]
G = tesla_img[:, :, 1]
R = tesla_img[:, :, 2]

# B, G, R print out arrays that refer to Blue, Green, and Red channel respectively
```

```python
# You can also convert image from BGR into RGB in a single instruction:

tesla_plt = tesla_img[:, :, ::-1]
```
